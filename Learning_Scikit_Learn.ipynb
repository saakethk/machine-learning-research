```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("aknjit/human-vs-ai-text-classification-dataset")
data_path = f"{path}/your_dataset_5000.csv"
```

```python
import pandas as pd
data = pd.read_csv(data_path)
data.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB

# Defines classifier
pipe = Pipeline([
    ("vect", CountVectorizer()),
    ("tfidf", TfidfTransformer()),
    ("clf", MultinomialNB())
])
```

```python
from sklearn.model_selection import train_test_split

# Creates train test split for data
TEST_RATIO = 0.2
RANDOM_SEED = 67
x_train, x_test, y_train, y_test = train_test_split(data["text"], data["label"], test_size=TEST_RATIO, random_state=RANDOM_SEED)
```

```python
# Fits model to pipeline
pipe.fit(x_train, y_train)
```

```python
# Generate predictions
predictions = pipe.predict(x_test)
print(predictions)
```

```python
from sklearn import metrics

# Get evaluation metrics
print(metrics.classification_report(y_test, predictions, target_names=["Is Human", "Is AI"]))
```
